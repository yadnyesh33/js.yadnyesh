Step 6:

Here, price is of object type(string), it should be int or float, so we need to change it

```python
data.price.unique()
```

```python




# Load the data (replace 'your_file.csv' with your actual file name)
data = pd.read_csv('output.csv')

# Clean column names (remove spaces and convert to lowercase)
data.columns = data.columns.str.strip().str.lower()

# Proceed only if 'price' column exists
if 'price' in data.columns:
    # Remove rows where 'price' is '?', ensuring a new copy is created
    data = data.loc[data['price'] != '?'].copy()

    # Convert 'price' column to numeric safely
    data.loc[:, 'price'] = pd.to_numeric(data['price'], errors='coerce')



    # Convert 'price' column to integer safely
    data.loc[:, 'price'] = data['price'].astype(int)

# Print the updated data types
print(data.dtypes)

# Display the first few rows to confirm changes
print(data.head())

```

Step 7:

Normalizing values by using simple feature scaling method examples(do for the rest) and binning- grouping values






```python
data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
data['height'] = data['height']/data['height'].max()

# binning- grouping values
bins = np.linspace(min(data['price']), max(data['price']), 4)
group_names = ['Low', 'Medium', 'High']
data['price-binned'] = pd.cut(data['price'], bins,
                              labels = group_names,
                              include_lowest = True)

print(data['price-binned'])
plt.hist(data['price-binned'])
plt.show()
```

Step 12:

Obtaining the final result and showing it in the form of a graph. As the slope is increasing in a positive direction, it is a positive linear relationship.

```python
# Select only 'make' and 'price' columns
data_annova = data[['make', 'price']]

# Group data by 'make'
grouped_annova = data_annova.groupby(['make'])

# Perform ANOVA test between 'honda' and 'subaru' prices
annova_results_l = sp.stats.f_oneway(
    grouped_annova.get_group(('honda',))['price'],
    grouped_annova.get_group(('subaru',))['price']
)

# Print results
print(annova_results_l)
```

```python
# Correlation analysis
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Outlier detection using Z-score and IQR
from scipy.stats import zscore

# Z-score method
z_scores = df.select_dtypes(include=['number']).apply(zscore)
df_no_outliers_z = df[(z_scores < 3).all(axis=1)]

# IQR method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df_no_outliers_iqr = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Shape after removing outliers using Z-score:", df_no_outliers_z.shape)
print("Shape after removing outliers using IQR:", df_no_outliers_iqr.shape)
```

```python
# Feature importance analysis using RandomForest
from sklearn.ensemble import RandomForestRegressor

# Assuming target variable is 'price'
X = df.drop(columns=['price'])
y = df['price']

# Training a simple model
model = RandomForestRegressor(random_state=42)
model.fit(X, y)

# Plotting feature importance
plt.figure(figsize=(10, 5))
sns.barplot(x=model.feature_importances_, y=X.columns)
plt.title("Feature Importance Analysis")
plt.show()
```

```python
# Custom functions for data preprocessing
def remove_duplicates(df):
    return df.drop_duplicates()

def detect_outliers_zscore(df):
    z_scores = df.select_dtypes(include=['number']).apply(zscore)
    return df[(z_scores < 3).all(axis=1)]

def detect_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Custom functions added for preprocessing.")
```
